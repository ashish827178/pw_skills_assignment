```python
def print_exception_hierarchy(base_exception, level=0):
    print('\t' * level, base_exception.__name__)
    for subclass in base_exception.__subclasses__():
        print_exception_hierarchy(subclass, level + 1)

print_exception_hierarchy(BaseException)
```

```python
# Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
# The ArithmeticError class in Python defines errors that occur during arithmetic operations. 
# Two errors defined in the ArithmeticError class are ZeroDivisionError and OverflowError.

# ZeroDivisionError: This error occurs when division or modulo by zero is attempted.
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print("Error:", e)
# OverflowError: This error occurs when the result of an arithmetic operation is too large to be represented.
try:
    result = 2 ** 1000000
except OverflowError as e:
    print("Error:", e)
```

```python
# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
# The LookupError class in Python is used to handle errors that occur when a key or index is not found in a mapping or sequence, 
# respectively. Two errors derived from LookupError are KeyError and IndexError.

# KeyError: This error occurs when a key is not found in a dictionary.
try:
    d = {'a': 1, 'b': 2}
    value = d['c']
except KeyError as e:
    print("Error:", e)
# IndexError: This error occurs when trying to access an index that is out of range in a sequence.
try:
    l = [1, 2, 3]
    value = l[4]
except IndexError as e:
    print("Error:", e)
```
